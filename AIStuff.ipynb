## Prompt: drop rows with null values

## AI: ChatGPT-5

```python
# Drop all rows with any null values
df = df.dropna()

# Confirm there are no more missing values
df.isnull().sum()
```

## Prompt: visualize the class distribution of a column named fraud

## AI: ChatGPT-5

```python
import matplotlib.pyplot as plt
import seaborn as sns
# Set a visual style
sns.set(style="whitegrid")

# Plot the class distribution
plt.figure(figsize=(6, 4))
sns.countplot(x='fraud', data=df, palette='Set2')

# Add labels and title
plt.title('Class Distribution of Fraud', fontsize=14)
plt.xlabel('Fraud')
plt.ylabel('Count')

# Show the plot
plt.show()
```

## Prompt: Now use a histplot with 30 bins for the distribution of the ratio_to_median_purchase_price column

## AI: ChatGPT-5

```python
# Set visual style
sns.set(style="whitegrid")

# Plot histogram for 'ratio_to_median_purchase_price'
plt.figure(figsize=(8, 5))
sns.histplot(data=df, x='ratio_to_median_purchase_price', bins=30, kde=True, color='skyblue')

# Add labels and title
plt.title('Distribution of Ratio to Median Purchase Price', fontsize=14)
plt.xlabel('Ratio to Median Purchase Price')
plt.ylabel('Frequency')

# Show the plot
plt.show()
```

## Prompt: create a coutplot of the column 'used_pin_number' with the hue being based on 'fraud' column with green and red palette

## AI: ChatGPT-5

```python
# Set visual style
sns.set(style="whitegrid")

# Plot countplot with hue based on 'fraud'
plt.figure(figsize=(6, 4))
sns.countplot(
    x='used_pin_number',
    hue='fraud',
    data=df,
    palette={0: 'green', 1: 'red'}
)

# Add labels and title
plt.title('Used PIN Number vs Fraud', fontsize=14)
plt.xlabel('Used PIN Number')
plt.ylabel('Count')
plt.legend(title='Fraud', loc='upper right')

# Show the plot
plt.show()
```

## Prompt: next create a bar plot based on the df grouped by 'used_pin_number' column for the 'fraud' column value counts with unstacking and converting the decimal point ot whole number percentage + and stacked

## AI: ChatGPT-5

```python
# Group by 'used_pin_number' and count fraud values
fraud_counts = df.groupby('used_pin_number')['fraud'].value_counts().unstack()

# Convert to percentage
fraud_percent = fraud_counts.div(fraud_counts.sum(axis=1), axis=0) * 100

# Plot as a stacked bar chart
fraud_percent.plot(kind='bar', stacked=True, color=['green', 'red'], figsize=(6, 4))

# Add labels and title
plt.title('Percentage of Fraud by Used PIN Number (Stacked)', fontsize=14)
plt.xlabel('Used PIN Number')
plt.ylabel('Percentage (%)')
plt.legend(title='Fraud', loc='upper right')

# Show the plot
plt.show()
```

## Prompt: lastly, create a scatterplot between the columns, 'distance_from_home' and 'ratio_to_median_purchase_price', with the hue as 'fraud' and green and red palette

## AI: ChatGPT-5

```python
# Set visual style
sns.set(style="whitegrid")

# Scatter plot
plt.figure(figsize=(8, 6))
sns.scatterplot(
    x='distance_from_home',
    y='ratio_to_median_purchase_price',
    hue='fraud',
    data=df,
    palette={0: 'green', 1: 'red'},
    alpha=0.6
)

# Add labels and title
plt.title('Distance from Home vs Ratio to Median Purchase Price', fontsize=14)
plt.xlabel('Distance from Home')
plt.ylabel('Ratio to Median Purchase Price')
plt.legend(title='Fraud', loc='upper right')

# Show the plot
plt.show()
```
